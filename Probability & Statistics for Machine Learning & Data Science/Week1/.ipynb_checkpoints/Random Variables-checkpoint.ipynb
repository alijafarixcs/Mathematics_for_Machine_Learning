```python
import matplotlib.pyplot as plt
import math
```

```python
def binomial_des(p,n,k):
    C=math.factorial(n)/(math.factorial(n-k)*math.factorial(k))
    #for example 5 times Dice to get 3 one is C(5,3) * 1/6**3 * 5/6 **2
    return C*(p**k)*((1-p)**(n-k))
n=20
p=0.113
dis_list=[]
for i in range(n+1):
    dis_list.append(binomial_des(p,n,i))
plt.bar([i for i in range(n+1)],dis_list)
```

```python
def random_generator(a, b, num_samples=100):
    import time
    ### START CODE HERE ###
    array =[]
    for i in range(num_samples):
        current_time = time.time()
        time.sleep(0.0001)
        # Get the fractional part of the current time
        fractional = current_time - int(current_time)
        fractional-=((fractional**3*i**2*current_time)%1)
        fractional=abs(fractional)
        x=fractional
        x=x*(b-a)+a
        array.append(x)
    
    ### END CODE HERE ###

    return array
un=random_generator(0,1,50)
plt.bar([i for i in range(len(un))],un)
```

```python
def uniform_generator(a, b,n):
    return [(a + (b-a))*r for r in random_generator(0,1,n)]
```

```python
def erfinv(x):
    a1 =  0.254829592
    a2 = -0.284496736
    a3 =  1.421413741
    a4 = -1.453152027
    a5 =  1.061405429
    p  =  0.3275911

    sign = 1 if x < 0 else -1
    x = abs(x)

    t = 1.0/(1.0 + p*x)
    y = (((((a5*t + a4)*t) + a3)*t + a2)*t + a1)*t
    return sign*(1 - y * math.exp(-x*x))

def inverse_cdf(p, mu=0, sigma=1):
    return mu + sigma * math.sqrt(2) * erfinv(2*p - 1)

def linespace(start,end,devid):
    w=end-start
    s=w/devid
    return [(start+(i*s)) for i in range(devid)]
def gaussian_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2 * math.pi)
    return (1.0 / (sqrt_two_pi * sigma)) * math.exp(-0.5 * ((x - mu) / sigma) ** 2)

def gaussian_generator(mu, sigma, num_samples):

    u = None
    u=uniform_generator(0,1,num_samples)
    
    array = None
    array=[inverse_cdf(y, mu, sigma) for y in u]
    

    return array
sigma=1
mu=0

X=linespace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(X,[gaussian_pdf(x,mu,sigma) for x in X])
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('pdf(x)')
plt.grid(True)
plt.show()
```
